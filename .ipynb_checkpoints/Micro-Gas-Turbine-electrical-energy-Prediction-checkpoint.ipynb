```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data
train_data = pd.read_csv('Dataset/Merged_dataset/merged_file.csv', names=['time', 'voltage', 'el_power'])
test_data = pd.read_csv('Dataset/Merged_dataset/test_merged_file.csv', names=['time', 'voltage'])
```

```python
# Convert 'time', 'voltage', and 'el_power' columns to float, invalid entries will be set to NaN
train_data['time'] = pd.to_numeric(train_data['time'], errors='coerce')
train_data['voltage'] = pd.to_numeric(train_data['voltage'], errors='coerce')
train_data['el_power'] = pd.to_numeric(train_data['el_power'], errors='coerce')

test_data['time'] = pd.to_numeric(test_data['time'], errors='coerce')
test_data['voltage'] = pd.to_numeric(test_data['voltage'], errors='coerce')
```

```python
# Handle missing values by dropping rows with NaN in the training data
train_data = train_data.dropna()
```

```python
# Check and print the number of NaN values in test_data
print("NaN values in test data:\n", test_data.isna().sum())
```

```python
# Fill missing values in test_data with some strategy (for now, forward fill, adjust as needed)
test_data = test_data.fillna(method='ffill')
```

```python
# Split features and target for training data
X_train = train_data[['time', 'voltage']]
y_train = train_data['el_power']
```

```python
# Split training data into training and validation sets (75:25 ratio)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
```

```python
# Check for NaN and infinity in target variables
print("Checking y_train for NaN or infinity:", np.any(np.isnan(y_train)), np.any(np.isinf(y_train)))
print("Checking y_val for NaN or infinity:", np.any(np.isnan(y_val)), np.any(np.isinf(y_val)))
```

```python
# Handle NaN or infinity values in y_train and y_val
y_train = np.nan_to_num(y_train, nan=np.nanmedian(y_train))
y_val = np.nan_to_num(y_val, nan=np.nanmedian(y_val))
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(test_data[['time', 'voltage']])
```

```python
# Initialize and train the XGBoost model
model = XGBRegressor(learning_rate=0.1, n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train, 
          eval_set=[(X_train_scaled, y_train), (X_val_scaled, y_val)],
          eval_metric='rmse',
          early_stopping_rounds=10,
          verbose=False)
```

```python
# Plot learning curves
train_scores = model.evals_result()['validation_0']['rmse']
val_scores = model.evals_result()['validation_1']['rmse']

plt.figure(figsize=(10, 6))
plt.plot(train_scores, label='Training RMSE')
plt.plot(val_scores, label='Validation RMSE')
plt.xlabel('Boosting Rounds')
plt.ylabel('RMSE')
plt.title('XGBoost Learning Curves')
plt.legend()
plt.show()
```

```python
# Make predictions on the validation set
y_val_pred = model.predict(X_val_scaled)
```

```python
# Calculate R2 score and RMSE for validation set
r2_val = r2_score(y_val, y_val_pred)
rmse_val = np.sqrt(mean_squared_error(y_val, y_val_pred))

print("Validation Set Metrics:")
print(f"Learning Rate: {model.learning_rate}")
print(f"R2 Score: {r2_val:.4f}")
print(f"RMSE: {rmse_val:.4f}")
```

```python
# Predict the values for the validation set
y_val_pred = model.predict(X_val_scaled)

# Plot predicted vs actual values for validation set
plt.figure(figsize=(12, 6))
plt.scatter(y_val, y_val_pred, alpha=0.5, color='blue', label='Predictions')
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'r--', lw=2, label='Ideal Fit')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predicted vs Actual Values (Validation Set)')
plt.legend()
plt.show()
```

```python
# Make predictions on the test set
test_data['el_power_predicted'] = model.predict(X_test_scaled)
```

```python
# Plot the predicted values for the test set
plt.figure(figsize=(12, 6))
plt.scatter(test_data['time'], test_data['el_power_predicted'], alpha=0.5)
plt.xlabel('Time')
plt.ylabel('Predicted Electric Power')
plt.title('Predicted Electric Power vs Time (Test Set)')
plt.show()
```
